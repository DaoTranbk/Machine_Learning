#### preprocessing 

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stop=set(stopwords.words('english'))
print(stop)
```

```python
lst=['won','nor','not','against']
for word in lst:
    stop.remove(word)
print(stop)
```

```python
sno=nltk.stem.SnowballStemmer('english')
```

```python
def cleanhtml(sentence):
    cleanr=re.compile('<.*?>')
    cleantext=re.sub(cleanr,' ',sentence)
    return cleantext
```

```python
def cleanpunc(sentence):
    '''This function cleans all the punctuation or special characters from a given sentence'''
    cleaned = re.sub(r'[?|@|!|^|%|\'|"|#]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    return  cleaned
```

```python
def preprocessing(series):
    i=0
    str1=" "
    final_string=[]
    list_of_sent=[]
    all_ham=[]
    all_spam=[]
    
    for sent in series.values:
        filtered_sent=[]
        list_of_sent=[]
        sent=cleanhtml(sent)
        sent=cleanpunc(sent)
        for cleaned_words in sent.split():
             if((cleaned_words.isalpha()) & (len(cleaned_words) > 2)):
                    if(cleaned_words.lower() not in stop):
                        s = (sno.stem(cleaned_words.lower()))
                        filtered_sent.append(s)    
                        if (email['class'].values)[i] == 0:
                            all_ham.append(s)
                        if(email['class'].values)[i] == 1:
                            all_spam.append(s)
        list_of_sent.append(filtered_sent)
        str1 = " ".join(filtered_sent)
        final_string.append(str1)
        i += 1
    return final_string, list_of_sent    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
conn = sqlite3.connect('email.sqlite')
email= pd.read_sql_query("""SELECT * FROM emails""",conn)
conn.close()
email
```

#### binomial naive bayes

```python
bow_vect=CountVectorizer(binary=True)
bow = bow_vect.fit_transform(email['Cleaned_email'].values)
```

```python
X=bow
y=email["class"].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,shuffle=False)
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB


from sklearn.metrics import accuracy_score
from sklearn.cross_validation import cross_val_score
```

```python
alpha= np.linspace(0.1,1,10)
cv_scores=[]

for a in alpha:
    clf=BernoulliNB(alpha=a)
    scores=cross_val_score(clf,X_train,y_train,cv=10,scoring='accuracy')
    cv_scores.append(scores.mean())

MSE=[1 - x for x in cv_scores]

optimal_alpha= alpha[MSE.index(min(MSE))]
print('\nThe optimal value of alpha is{}'.format(optimal_alpha))

plt.plot(alpha,MSE)
plt.xlabel('Alpha')
plt.ylabel('Misclassification error')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
clf=BernoulliNB(alpha=optimal_alpha)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
acc=accuracy_score(y_test,y_pred,normalize=True)*float(100)
print(acc)
confusion_matrix(y_test,y_pred).T
```

```python
idx=(-clf.feature_log_prob_).argsort()[:2,:50]
print('Top 50 words in negative reviews')
for i in idx[0]:
    print(bow_vect.get_feature_names()[i])
    
print("***********")
print('Top 50 words in postive reviews')
for i in idx[1]:
    print(bow_vect.get_feature_names()[i])
```
